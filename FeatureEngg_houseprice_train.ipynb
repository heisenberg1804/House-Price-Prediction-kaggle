## Null Values

```python
#EDA- Nan values
cat_with_na = [var for var in df_train.columns if df_train[var].isnull().sum() > 1 and df_train[var].dtypes == 'O']
print(cat_with_na)
df_train[cat_with_na].isnull().mean()
```

## Replacing NA values in categorical features with 'Missing'

```python
def replace_cat_with_missing(dataset, cat_with_na):
    data = dataset.copy()
    data[cat_with_na] = data[cat_with_na].fillna('Missing')
    return data

df_train = replace_cat_with_missing(df_train, cat_with_na)
df_train[cat_with_na].isnull().mean()
```

```python
df_train.shape
```

```python
num_with_na = [var for var in df_train.columns if df_train[var].isnull().sum() > 0 and df_train[var].dtypes != 'O']
print(num_with_na)
df_train[num_with_na].isnull().mean()
```

## Replacing NA values in numerical features with median because of outliers

```python
for var in num_with_na:
    #calculate median because there are a lot of outliers
    median_value = df_train[var].median()
    
    df_train[var].fillna(median_value, inplace=True)

df_train[num_with_na].isnull().mean()
```

```python
year_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']
```

## Changing 'YearBuilt', 'YearRemodAdd', 'GarageYrBlt' to their difference between 'YrSold'

```python
for feature in year_features:
    if feature != 'YrSold':
        df_train[feature] = df_train['YrSold'] - df_train[feature]
       
```

## Now converting Training data to X_train and Y_train for training

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout


# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 50, init = 'he_uniform',activation='relu',input_dim = 137))

# Adding the second hidden layer
classifier.add(Dense(output_dim = 25, init = 'he_uniform',activation='relu'))

# Adding the third hidden layer
classifier.add(Dense(output_dim = 50, init = 'he_uniform',activation='relu'))
# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'he_uniform'))

# Compiling the ANN
classifier.compile(loss=keras.losses.MeanSquaredError(), optimizer='Adamax')

# Fitting the ANN to the Training set
model_history=classifier.fit(X_train.values, y_train.values,validation_split=0.20, batch_size = 10, nb_epoch = 1000)
```
